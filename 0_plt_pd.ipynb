````python
import matplotlib.pyplot as plt
import numpy as np
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, marker='o')
plt.title("Basic Components of Matplotlib Figure")
plt.xlabel("X-Axis") 
plt.ylabel("Y-Axis")  
plt.grid(True)
plt.show() 
````

````python
fig, ax = plt.subplots()  
ax.plot(x, y, marker='o', label="Data Points")

ax.set_title("Basic Components of Matplotlib Figure")
ax.set_xlabel("X-Axis") 
ax.set_ylabel("Y-Axis")  


plt.show()
````

This is Matplotlib’s **object-oriented style** of plotting.

What each line does:

```python
fig, ax = plt.subplots()  
ax.plot(x, y, marker='o', label="Data Points")

ax.set_title("Basic Components of Matplotlib Figure")
ax.set_xlabel("X-Axis") 
ax.set_ylabel("Y-Axis")  

plt.show()
```

1. `fig, ax = plt.subplots()`
- Creates:
  - `fig` = the **Figure** (the whole canvas/window).
  - `ax` = an **Axes** object (the actual plot area inside the figure).

2. `ax.plot(...)`
- Draws the line on that specific axes (`ax`), with circle markers.

3. `ax.set_title(...)`, `ax.set_xlabel(...)`, `ax.set_ylabel(...)`
- Sets title and axis labels for that same axes.

4. `plt.show()`
- Displays the figure on screen.

Why use `fig, ax`?
- It gives you **explicit control** over where things are drawn.
- It scales well for multiple plots, e.g. `fig, axs = plt.subplots(2, 2)`.
- It avoids confusion from the global `plt` state (cleaner and more maintainable code).

In short:
- `fig` = whole page
- `ax` = one chart on that page


````python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].plot(x, y)
axs[0].set_title("Left plot")

axs[1].bar(x, y)
axs[1].set_title("Right plot")

plt.tight_layout()
plt.show()
````

````python
# Line2D is a low-level Matplotlib object. Creating it alone won’t 
# draw anything—you have to add it to an Axes.
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

x = [0, 1, 2, 3]
y = [0, 1, 4, 9]

fig, ax = plt.subplots()

line = Line2D(x, y, color='blue', linewidth=2)
ax.add_line(line)

ax.set_xlim(min(x), max(x))
ax.set_ylim(min(y), max(y))

plt.show()
````

````python
from sklearn.linear_model import Perceptron
from sklearn.datasets import load_iris
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load and split data
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

# Initialize and train model
model = Perceptron(random_state=42, max_iter=1000, tol=1e-3)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))   
# Generate confusion matrix
class_names = iris.target_names
disp = ConfusionMatrixDisplay.from_estimator(
    model,
    X_test,
    y_test,
    display_labels=class_names,
    # cmap=plt.cm.Blues,
    normalize=None  # Set to 'true' for normalized matrix
)

disp.ax_.set_title("Confusion Matrix (Without Normalization)")
plt.show()  
````

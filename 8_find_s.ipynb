```python
training_data = [
    # Sky Temp Humidity Wind Water Forecast Output
    ['Sunny','Warm','Normal','Strong','Warm','Same','Yes'],
    ['Sunny','Warm','High','Strong','Warm','Same','Yes'],
    ['Rainy','Cold','High','Strong','Warm','Change','No'],
    ['Sunny','Warm','High','Strong','Cool','Change','Yes'],
]
# Initialize hypothesis with first positive example
hypothesis = None
for sample in training_data:
    if sample[-1] == 'Yes':
        if hypothesis is None:
            hypothesis = sample[:-1].copy()
        else:
            for i in range(len(hypothesis)):
                if hypothesis[i] != sample[i]:
                    hypothesis[i] = '?' 
                    
print("Most Specific Hypothesis:")
print(hypothesis)
```

```python
import numpy as np

def find_s(concepts, target):
    # Initialize the specific hypothesis with the first positive example
    for i, val in enumerate(target):
        if val == "Yes":
            specific_h = concepts[i].copy()
            break
    
    # Process the remaining examples
    for i, h in enumerate(concepts):
        if target[i] == "Yes":
            for x in range(len(specific_h)):
                # If the attribute value differs, replace it with a wildcard '?'
                if h[x] != specific_h[x]:
                    specific_h[x] = '?'
        # Negative examples are completely ignored in Find-S
                    
    return specific_h

# --- Example Usage ---
data = np.array([
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Rainy', 'Cold', 'High', 'Strong', 'Warm', 'Change', 'No'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change', 'Yes']
])

concepts = data[:, :-1]
target = data[:, -1]

result = find_s(concepts, target)
print(f"The most specific hypothesis is: {result}")
```

````python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
import numpy as np

X, _ = make_blobs(n_samples=150, centers=3, cluster_std=0.6, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
print("Cluster Centers:")

for i, c in enumerate(centers):
    print(f" Cluster {i}: ({c[0]:.2f}, {c[1]:.2f})")

print("Inertia :", round(kmeans.inertia_, 2))
print("Silhouette Score:", round(silhouette_score(X, labels), 4))
````

````python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=labels, alpha=0.7)
plt.scatter(centers[:, 0], centers[:, 1],
            marker='X', s=200, c='black', label='Centroids')

plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Sample dataset (2D points)
X = np.array([
    [1, 2], [1.5, 1.8], [5, 8],
    [8, 8], [1, 0.6], [9, 11],
    [8, 2], [10, 2], [9, 3]
])

# Create KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit model
kmeans.fit(X)

# Get cluster labels
labels = kmeans.labels_

# Get centroids
centroids = kmeans.cluster_centers_

print("Cluster Labels:", labels)
print("Centroids:\n", centroids)

# Visualization
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='X', s=200)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("K-Means Clustering")
plt.grid(True)
plt.show()
````


---

## 📊 What This Code Does

✔ Creates sample data
✔ Applies K-Means algorithm
✔ Divides data into **3 clusters**
✔ Finds **centroids**
✔ Visualizes clusters

---

## 🔍 Output Explanation

* `labels` → cluster assigned to each point
* `centroids` → center of each cluster
* Graph:

  * Different colors → different clusters
  * `X` marks → centroids

---

## 🧠 How K-Means Works (Simple Steps)

1. Choose number of clusters (K)
2. Initialize centroids
3. Assign points to nearest centroid
4. Update centroids
5. Repeat until stable

---

## ⚠️ Important Points

* K must be chosen manually
* Works best for **well-separated clusters**
* Sensitive to initial centroid placement

---

## 🎓 Viva Questions

### 1️⃣ What is K-Means?

An unsupervised learning algorithm used to group data into K clusters.

### 2️⃣ What is centroid?

The mean position of all points in a cluster.

### 3️⃣ What does `n_clusters` do?

Specifies number of clusters (K).

---

## 📌 Exam-Ready Definition

> K-Means is an unsupervised learning algorithm that partitions data into K clusters by minimizing the distance between data points and their cluster centroids.


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Load dataset
iris = load_iris()
X = iris.data[:, [2, 3]]   # Use petal length & width (best for clustering)

# Create KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit model
kmeans.fit(X)

# Get cluster labels
labels = kmeans.labels_

# Get centroids
centroids = kmeans.cluster_centers_

print("Centroids:\n", centroids)

# Visualization
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='X', s=200)

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("K-Means Clustering on Iris Dataset")
plt.grid(True)
plt.show()
````
